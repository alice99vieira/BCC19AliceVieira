1) Faça uma função que recebe dois vetores ou duas colunas de uma DataFrame do Pandas e retorne a correlação . A função deve estar no mesmo arquivo que a função que calcula a regressão. Esse arquivo deve ser importado no script que executará o teste abaixo.

Teste com o dado de público e o dado da renda dos jogos do campeonato brasileiro de 2018 e mostre a correlação encontrada.


```python
import pandas as pd
TB = pd.read_csv('https://raw.githubusercontent.com/alice99vieira/BasesComputacionais/bacd478ebcde398e853ef5c700168b98f105d3a3/aula9/dados/tabelaBrasileirao2018.csv')
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 7))
plt.plot(TB['Público'], TB['Renda (R$)'], marker = 'o', linestyle = '', 
         color = 'red', markersize =7)
plt.grid()
plt.xlabel('Público')
plt.ylabel('Renda (R$)')
plt.show()
```

```python
corr_Público_Renda = TB['Público'].corr(TB['Renda (R$)'])
print(corr_Público_Renda)
```
