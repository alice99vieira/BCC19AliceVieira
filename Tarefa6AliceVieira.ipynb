1) A tensão elétrica medida em um circuito, em volts, tem a seguinte expressão no instante t, em segundos:

Faça o gráfico da tensão em função do tempo.

Sugestão: fazer o gráfico com tempo entre 0 e 10 s com incremento de 0.05 s.

```python
import matplotlib.pyplot as plt
import numpy as np
t = np.array([0, 1.5, 1,55, 2, 3.05, 3,5, 4, 4.75, 5.70, 7.05, 9,55, 10])
vt = np.exp(-0.5*t)*np.cos(2*(np.pi)*0.8*t)
vt

```

```python
plt.figure()
plt.xlabel('t(segundos)')
plt.ylabel('tensão(volts)')
plt.title('A tensão elétrica medida em um circuito')
plt.plot(t, vt, marker='d', color='red', linestyle ='-.')
plt.grid()
plt.show()
```

3) Entre os anos de 1994 e 2022, a população de uma cidade cresce aproximadamente de acordo com a seguinte função:
 
em que a é o ano e P é o número de habitantes.

Faça o gráfico do número de habitantes entre 1994 e 2022.

```python
import matplotlib.pyplot as plt
import numpy as np
a = np.linspace(1994, 2022, 100)
a
Pa = 800000*np.exp((a-1994)/40)
Pa
```

```python
plt.figure()
plt.xlabel('anos(a)')
plt.ylabel('número de habitante(p)')
plt.title('habitantes entre 1994 e 2022')
plt.plot(a, Pa, color='blue', linestyle ='-')
plt.grid()
plt.show()
```
