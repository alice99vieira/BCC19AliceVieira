1) Faça a aproximação de pi usando a fórmula de Leibniz:

Calcule (e mostre na tela) a aproximação de  considerando a) 50 termos, b) 500 termos, c) 1000 termos e d) 10000 termos.

Dica: Note que o sinal do termo muda a cada termo. Talvez verificar se o  é par ou ímpar ajude a decidir se para aquele i o sinal é mais ou menos.



a) 50 termos

```python
N = 50
pi = 0
for i in range(1,N*2 + 1,2):
  S = -(-1)**((i+1)/2)
  pi += 4 / i*S
  print(f'S = {pi}')
```
